```python
import numpy as np 
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
from IPython.display import Image, display
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read data
data = pd.read_csv("telecom_churn.csv")
data.head()
```

```python
data.info()
```

### Check the target variable

Amount of churned customers
How many customers have churned?

```python
feature = "Churn"
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
data[feature].value_counts().plot.pie(autopct='%1.1f%%')
plt.subplot(1,2,2)
sns.countplot(data[feature])
plt.show()
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1,3 )
f.set_figheight(8)
f.set_figwidth(15)

sns.kdeplot(data['Total intl charge'], hue=data.Churn, ax=ax1)
sns.kdeplot(data['Total day charge'], hue=data.Churn, ax=ax2)
sns.kdeplot(data['Total eve charge'], hue=data.Churn, ax=ax3);
```
